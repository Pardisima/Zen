**1) Trabajemos con NumPy**

* Importa numpy y usa un alias para poder utilizar sus recursos
* Crea un array de elementos (del 10 al 100)
* Invertir el array
* Crear un array 3x3 con valores de 0 a 8
* Crear un array identidad de 6x6
* Crear un array con valores al azar con forma 3x3x3
* Crear dos arrays con valores al azar A y B, verificar si son iguales (hay varias opciones, investigar)

```python
# 1) Importo numpy dandole un alias
import numpy as np

# Creo un array de elementos del 10 al 100 usando la funcion "np.arange"
miPrimerArray = np.arange(10,101)
print("Array de elementos del 10 al 100:\n\n",miPrimerArray)

# Invierto el array
array_inverso = np.flip(miPrimerArray)
print("\nArray invertido:\n\n",array_inverso)

# Creo un array de 3x3 con valores de 0 a 8
mi_primera_matriz = np.random.choice(10, (3, 3))
print("\nArray de 3x3 con valores de 0 a 8:\n\n",mi_primera_matriz)

# Creo un array identidad de 6x6
array_identidad = np.identity(6)
print("\nArray de identidad de 6x6:\n\n",array_identidad)

# Creo un array con valores al azar con forma 3x3x3
mi_matriz_3D = np.random.rand(3,3,3)
print("\nArray con valores al azar con forma de 3x3x3:\n\n",mi_matriz_3D)

# Creo dos arrays con valores al azar A y B y verifico si son iguales
print("\nCreo dos arrays con valores al azar A y B\n")
A = np.random.rand(5)
print("Array A: ",A)
B = np.random.rand(5)
print("Array B: ",B)
C = A == B
print("\nVerifico si los valores de A son iguales a los valores de B:\n\n",C)
```

**2) Sigamos con NumPy**

* Crear un array de 100 elementos aleatorios comprendidos entre 1 y 1000 (pueden ser números enteros).

* Realizar la conversión de los precios en dólares del primer array al segundo array en euros. Se puede serguir la siguiente conversión:
    - 1 EUR = 1,14362 USD

```python
# 2) Creo un array llamado dólares de 100 elementos aleatorios comprendidos entre el 1 y el 1000
dólares = np.random.randint(1,1001, 100)
print("Muestra del array 'dólares'\n\n",dólares)
# Creo otro array llamado euros
print("\n Muestra del array 'euros' redondeado a 2 decimales\n")
euros = (dólares * 0.874413)
print(euros.round(2))
```

```python
# 3) Importo el modulo "pandas" y le doy el alias "pd"
import pandas as pd
# Cargo como dataframe el csv 05_05_imbd_titulos.csv y muestro sus primeros 5 registros
imbd_titulos = pd.read_csv("05_05_imdb_titulos.csv")
print("Los primeros 5 registros de imbd_titulos:\n")
imbd_titulos.head()
```

```python
# Cargo como dataframe el csv 05_05_imbd_elenco.csv y muestro sus primeros 5 registros
imbd_elenco = pd.read_csv("05_05_imdb_elenco.csv")
print("\nLos primeros 5 registros de imbd_elenco:\n")
imbd_elenco.head()
```

```python
# Muestro el numero de registros de 'titulos'
print("Cantidad de registros de títulos: --> ",len(imbd_titulos["title"]), "registros")

# Muestro el numero de registros de 'elenco'
print("\nCantidad de registros de elenco: --> ",len(imbd_elenco["title"]), "registros")
```

```python
# Muestro las 5 peliculas más antiguas del listado de titulos
titulos_ordenados_anio = imbd_titulos.sort_values(by="year",ascending=False)
print("Las 5 películas más antiguas del listado de títulos:")
titulos_ordenados_anio[-5:]
```

```python
# Muestro las peliculas con titulo "Dracula" y el número total de peliculas que coincidan con el requisito 
titulos_dracula = imbd_titulos["title"][ imbd_titulos["title"] == "Dracula"]
print("Muestro todas las películas con título 'Dracula':\n")
print(titulos_dracula)
print("\nNúmero total de películas que coinciden con 'Dracula' ---> ",len(titulos_dracula),"\n")
```

```python
# Muestro los 10 titulos mas comunes
df_title = imbd_titulos.groupby("title") # Agrupo los titulos
df_title = df_title.size() # Agrupo cada titulo con el numero de coincidencias
df_title_rep = df_title.sort_values(ascending=False) # Ordeno por numero de coincidencias
print("Lista de los 10 títulos más comunes: \n")
print(df_title_rep.head(10))
# Muestro la primera pelicula de "Romeo and Juliet"
imbd_titulos_group = imbd_titulos.groupby("title")
romeo_juliet = imbd_titulos_group.get_group("Romeo and Juliet")["year"].min()
print("\nLa primera película de 'Romeo y Julieta' es del año ",romeo_juliet)
```

```python
# Listo todas las peliculas que contengan la palabra "Exorcist" ordenadas de la más vieja a la más reciente
print("\nListo todas las peliculas que contengan la palabra 'Exorcist' ordenadas de la más vieja a la más reciente\n")
imbd_titulos[imbd_titulos.title.str.contains("Exorcist")].sort_values('year')
```

```python
# Muestro cuantas peliculas son del año 1950
print("\nPelículas del año 1950:\n")
imbd_titulos.loc[ imbd_titulos["year"] == 1950, : ]
```

```python
# Muestro cuantas peliculas son entre el año 1950 y el 1959
print("\nPelículas entre el año 1950 y el 1959:\n")
imbd_titulos.loc[( imbd_titulos["year"] >= 1950) & (imbd_titulos["year"] < 1960), : ]
```

```python
# Muestro todos los roles  que hubo en la pelicula "The Godfather"
print("\nRoles de la película 'The Godfather'\n")
roles_elPadrino = imbd_elenco.loc[ imbd_elenco["title"] == "The Godfather", : ]
print(roles_elPadrino["character"])

# Muestro el número total de coincidencias
print("\nCoincidencias en roles:\n")
roles_elPadrino = imbd_elenco.loc[ imbd_elenco["title"] == "The Godfather", : ]
roles_elPadrino = roles_elPadrino.groupby("character")
print(roles_elPadrino.size())
```

```python
# Muestro cuantos papeles de "Bruce Wayne" han sido hechos en la historia de las peliculas
imbd_elenco.loc[ imbd_elenco["character"] == "Bruce Wayne", : ]
```

```python
# Muestro cuantos papeles ha hecho "Robert De Niro" en su carrera
imbd_elenco.loc[ imbd_elenco["name"] == "Robert De Niro", : ]
```

```python
# Listado de papeles protagonistas que tuvo el actor "Charlton Heston" en la década de los 60's ordenado por año de forma desc
charlton = imbd_elenco.loc[ (imbd_elenco["name"] == "Charlton Heston") & (imbd_elenco["year"] < 1970) & (imbd_elenco["year"] >= 1960) & (imbd_elenco["n"] >= 1), : ]
charlton.sort_values(by="year",ascending=False)
```

```python
# Mostrar cuantos papeles para actores hubo en la década de los 50's
numActores = len(imbd_elenco.loc[ (imbd_elenco["type"] == "actor") & (imbd_elenco["year"] < 1960) & (imbd_elenco["year"] >= 1950), : ])
print("El número de actores en la década de los 50's es: --->",numActores)
imbd_elenco.loc[ (imbd_elenco["type"] == "actor") & (imbd_elenco["year"] < 1960) & (imbd_elenco["year"] >= 1950), : ]
```

```python
# Mostrar cuantos papeles para actrices hubo en la década de los 50's
numActrices = len(imbd_elenco.loc[ (imbd_elenco["type"] == "actress") & (imbd_elenco["year"] < 1960) & (imbd_elenco["year"] >= 1950), : ])
print("El número de actrices en la década de los 50's es: --->",numActrices)
imbd_elenco.loc[ (imbd_elenco["type"] == "actress") & (imbd_elenco["year"] < 1960) & (imbd_elenco["year"] >= 1950), : ]
```
