## Creando una jerarquía de productos con clases



### Superclase Producto

```python
class Producto():
    def __init__(self,idproducto,nombre,precio,descripcion):
        self.idproducto = idproducto
        self.nombre = nombre
        self.precio = precio
        self.descripcion = descripcion
        print("Constructor de la clase Producto")
        
    def __str__(self):
        return """-> PRODUCTO <-
        - IdProducto: {}
        - Nombre: {}
        - Precio: {}
        - Descripción: {}""".format(self.idproducto,self.nombre,self.precio,self.descripcion)
```

### Subclase Alimento

```python
class Alimento(Producto):
    productor = ""
    distribuidor = ""
    
    def __str__(self):
        return """-> PRODUCTO ALIMENTO <-
        - IdProducto: {}
        - Nombre: {}
        - Precio: {}
        - Descripción: {}
        - Productor: {}
        - Distribuidor: {}""".format(self.idproducto,self.nombre,self.precio,self.descripcion,
                                     self.productor,self.distribuidor)
        
    
al = Alimento(48975,"Botella de Aceite de Oliva Extra",5,"250 ML")
al.productor = "La Aceitera"
al.distribuidor = "Distribuciones SA"

print(al)
```

# Pongamoslo todo junto y probemos algunas cosas

```python
class Producto():
    def __init__(self,idproducto,nombre,precio,descripcion):
        self.idproducto = idproducto
        self.nombre = nombre
        self.precio = precio
        self.descripcion = descripcion
        print("Constructor de la clase Producto")
        
    def __str__(self):
        return """-> PRODUCTO <-
        - IdProducto: {}
        - Nombre: {}
        - Precio: {}
        - Descripción: {}""".format(self.idproducto,self.nombre,self.precio,self.descripcion)
    
class Alimento(Producto):
    productor = ""
    distribuidor = ""
    
    def __str__(self):
        return """-> PRODUCTO ALIMENTO <-
        - IdProducto: {}
        - Nombre: {}
        - Precio: {}
        - Descripción: {}
        - Productor: {}
        - Distribuidor: {}""".format(self.idproducto,self.nombre,self.precio,self.descripcion,self.productor,self.distribuidor)
    
class Libro(Producto):
    isbn = ""
    autor = ""
    
    def __str__(self):
        return """-> PRODUCTO LIBRO <-
        - IdProducto: {}
        - Nombre: {}
        - Precio: {}
        - Descripción: {}
        - ISBN: {}
        - Autor: {}""".format(self.idproducto,self.nombre,self.precio,self.descripcion,self.isbn,self.autor)
    
class Ropa(Producto):
    talla = ""
    marca = ""
    
    def __str__(self):
        return """-> PRODUCTO ROPA <-
        - IdProducto: {}
        - Nombre: {}
        - Precio: {}
        - Descripción: {}
        - Talla: {}
        - Marca: {}""".format(self.idproducto,self.nombre,self.precio,self.descripcion,self.talla,self.marca)
```

### Creación de objetos

```python
al = Alimento(48975,"Botella de Aceite de Oliva Extra",5,"250 ML")
al.productor = "La Aceitera"
al.distribuidor = "Distribuciones SA"

li = Libro(15897,"Cocina Japonesa",15,"Recetas de oriente")
li.isbn = "1265626526856"
li.autor = "Maestro Sun"

ro = Ropa(11236,"Jersey Fit18",18,"Jersey de hombre de invierno")
ro.talla = "36"
ro.marca = "Guci"
```

### Lista de productos

```python
productos = [al, li, ro]
print(productos)
```

### Lectura de productos

```python
for p in productos:
    print(p,"\n")
```

#### Podemos acceder a los atributos si son compartidos entre todos los objetos

```python
for p in productos:
    print(p.idproducto, "->", p.nombre)
```

#### Pero si un objeto no tiene el atributo deseado, dará error:

```python
for p in productos:
    print(p.autor)
```

#### Tendremos que tratar cada subclase de forma distinta, gracias a la función isistance():

```python
for p in productos:
    if( isinstance(p, Alimento) ):
        print(p.idproducto, "->", p.nombre, "->", p.productor)
    elif( isinstance(p, Libro) ):
        print(p.idproducto, "->", p.nombre, "->", p.isbn)     
    elif( isinstance(p, Ropa) ):
        print(p.idproducto, "->", p.nombre, "->", p.marca)
```

## Uso de super()
Hasta ahora, hemos definido las clases hijas sin constructor, de tal modo, cuando instanciamos un objeto de estas clases hijas, usan el constructor del padre para construirse. Este mecánismo puede estar bien cuando nuestras clases hijas no tienen atributos especificos o tienen muy pocos, y su forma de crearse es muy similar a la del padre. Pero, ¿qué pasa cuando una clase hija tiene muchos atributos propios o su forma de crearse difiere mucho de la forma de su padre? En este caso tenemos que utilizar el método especial __super()__
<br><br>El método super() invocado desde una clase hija, hace referencia al padre. Por lo tanto, super() solo funciona cuando tenemos una estructura de herencia.
<br><br>Podemos utilizar super() con el siguiente formato: <code>__super().cualquier_metodo_de_la_clase_padre()__</code>
<br><br>Veámos un ejemplo a continuación, la clase Alimento que hereda de Producto tiene su propio constructor, donde realiza dos tareas:
* Definir sus atributos propios (productor y distribuidor) .
* Invocar al constructor del padre con ``super().__init__()`` y enviarle por parámetro los atributos comúnes que pertenecen a la clase del padre (referencia, nombre, pvp y descripción).

**Es muy importante que el número y orden de los parámetros que enviamos desde el hijo (con** ``super().__init__()``) **al padre, sean los mismos en número y orden que los que hay definidos en el constructor del padre.**

```python
class Producto:
    def __init__(self,idproducto=None,nombre=None,precio=None,descripcion=None):
        self.idproducto = idproducto
        self.nombre = nombre
        self.precio = precio
        self.descripcion = descripcion
        print("Constructor de la clase Producto")
        
    def saludar(self):
        print("Buenos dias")

class Alimento(Producto):
    def __init__(self,idproducto=None,nombre=None,precio=None,descripcion=None,
                 productor=None,distribuidor=None):
        super().saludar()
        super().__init__(idproducto,nombre,precio,descripcion)
        self.productor = productor
        self.distribuidor = distribuidor
        print("Constructor de la clase Alimento")

alimento1 = Alimento(idproducto=2035, nombre="Aceite Oliva Extra", precio=5, 
                     descripcion="Botella de 1l de aceite", productor="La Aceitera", 
                     distribuidor="Distribuciones SL")
alimento2 = Alimento(idproducto=2036, nombre="Vinagre", descripcion="Vinagre extra", 
                     productor="La Vinagrera", distribuidor="Distribuciones SL")

print(alimento1.nombre)
print(alimento2.nombre)
if alimento1.precio is not None and alimento2.precio is not None:
    precioTotal = alimento1.precio + alimento2.precio # Esto puede dar error porque suma 5 + None
else:
    print(" > Error en precios")
```
