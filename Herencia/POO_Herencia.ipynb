## Uso de super()
Hasta ahora, hemos definido las clases hijas sin constructor, de tal modo, cuando instanciamos un objeto de estas clases hijas, usan el constructor del padre para construirse. Este mec치nismo puede estar bien cuando nuestras clases hijas no tienen atributos especificos o tienen muy pocos, y su forma de crearse es muy similar a la del padre. Pero, 쯤u칠 pasa cuando una clase hija tiene muchos atributos propios o su forma de crearse difiere mucho de la forma de su padre? En este caso tenemos que utilizar el m칠todo especial __super()__
<br><br>El m칠todo super() invocado desde una clase hija, hace referencia al padre. Por lo tanto, super() solo funciona cuando tenemos una estructura de herencia.
<br><br>Podemos utilizar super() con el siguiente formato: <code>__super().cualquier_metodo_de_la_clase_padre()__</code>
<br><br>Ve치mos un ejemplo a continuaci칩n, la clase Alimento que hereda de Producto tiene su propio constructor, donde realiza dos tareas:
* Definir sus atributos propios (productor y distribuidor) .
* Invocar al constructor del padre con ``super().__init__()`` y enviarle por par치metro los atributos com칰nes que pertenecen a la clase del padre (referencia, nombre, pvp y descripci칩n).

**Es muy importante que el n칰mero y orden de los par치metros que enviamos desde el hijo (con** ``super().__init__()``) **al padre, sean los mismos en n칰mero y orden que los que hay definidos en el constructor del padre.**

```python
class Producto:
    def __init__(self,idproducto=None,nombre=None,precio=None,descripcion=None):
        self.idproducto = idproducto
        self.nombre = nombre
        self.precio = precio
        self.descripcion = descripcion
        print("Constructor de la clase Producto")
        
    def saludar(self):
        print("Buenos dias")

class Alimento(Producto):
    def __init__(self,idproducto=None,nombre=None,precio=None,descripcion=None,
                 productor=None,distribuidor=None):
        super().saludar()
        super().__init__(idproducto,nombre,precio,descripcion)
        self.productor = productor
        self.distribuidor = distribuidor
        print("Constructor de la clase Alimento")

alimento1 = Alimento(idproducto=2035, nombre="Aceite Oliva Extra", precio=5, 
                     descripcion="Botella de 1l de aceite", productor="La Aceitera", 
                     distribuidor="Distribuciones SL")
alimento2 = Alimento(idproducto=2036, nombre="Vinagre", descripcion="Vinagre extra", 
                     productor="La Vinagrera", distribuidor="Distribuciones SL")

print(alimento1.nombre)
print(alimento2.nombre)
if alimento1.precio is not None and alimento2.precio is not None:
    precioTotal = alimento1.precio + alimento2.precio # Esto puede dar error porque suma 5 + None
else:
    print(" > Error en precios")
```
